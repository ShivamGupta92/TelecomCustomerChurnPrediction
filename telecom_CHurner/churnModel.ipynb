```python
df.drop('customerID', axis='columns',inplace=True)
```

```python
df1 = df[df.TotalCharges!=' ']
```

```python
df1.shape
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges, errors='coerce')
```

```python
def print_unique_col_values(df):
    for coloumn in df:
        if df[coloumn].dtypes == 'object':
            print(f'{coloumn} : {df[coloumn].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No phone service', 'No', inplace=True)
df1.replace('No internet service', 'No', inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
object_columns = df1.select_dtypes(include=['object']).columns 
for i in object_columns:
    print(i)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
df1['gender']= label_encoder.fit_transform(df1['gender'])
df1['Dependents']= label_encoder.fit_transform(df1['Dependents'])
df1['PhoneService']= label_encoder.fit_transform(df1['PhoneService'])
df1['MultipleLines']= label_encoder.fit_transform(df1['MultipleLines'])
df1['InternetService']= label_encoder.fit_transform(df1['InternetService'])
df1['OnlineSecurity']= label_encoder.fit_transform(df1['OnlineSecurity'])
df1['OnlineBackup']= label_encoder.fit_transform(df1['OnlineBackup'])
df1['DeviceProtection']= label_encoder.fit_transform(df1['DeviceProtection'])
df1['TechSupport']= label_encoder.fit_transform(df1['TechSupport'])
df1['StreamingTV']= label_encoder.fit_transform(df1['DeviceProtection'])
df1['StreamingMovies']= label_encoder.fit_transform(df1['StreamingTV'])
df1['Contract']= label_encoder.fit_transform(df1['Contract'])
df1['PaperlessBilling']= label_encoder.fit_transform(df1['PaperlessBilling'])
df1['PaymentMethod']= label_encoder.fit_transform(df1['PaymentMethod'])
df1['Partner']= label_encoder.fit_transform(df1['Partner'])
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df1[to_scale] = scaler.fit_transform(df1[to_scale])
```

```python
df1.sample(5)
```

```python
df1.reset_index(drop = True, inplace = True)
df1
```

```python
X = df1.drop(["Churn"], axis=1)
Y = df1["Churn"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(20, kernel_initializer = 'he_normal', activation = 'relu', input_dim = 19),
    keras.layers.Dense(15, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dense(10, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dense(1, kernel_initializer = 'glorot_uniform', activation= 'sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(
    loss='binary_crossentropy',
    metrics=['accuracy'],
    optimizer='adam'
)
```

```python
model.fit(x_train, y_train, batch_size=4, epochs=50,verbose=1)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp = model.predict(x_test)
yp[:5]
```

```python
y_pred = []
for elem in yp:
    if elem>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_test[:10]
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions= y_pred)

plt.figure(figsize=(18,7))
sns.heatmap(cm,annot=True, fmt='d')
plt.xlabel('Predicition')
plt.ylabel('Truth')
```
